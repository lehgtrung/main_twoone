```python
import pandas as pd
import numpy as np
```

```python
import json
```

```python
with open('indices.json', 'r') as f:
    indices = json.load(f)
```

```python
indices[:5]
```

```python
with open('unlabeled.json', 'r') as f:
    unlabeled = json.load(f)
```

```python
no_match = 0
for i in indices:
    pred = [tuple(e) for e in unlabeled[i]['entities']]
    gt = unlabeled[i]['entity_gts']
    gt = [tuple(e[:3]) for e in gt]
    print(pred)
    print(gt)
    if set(pred) != set(gt):
        no_match += 1
    print('========================')
```

```python
no_match
```

```python
no_match = 0
for i in indices:
    pred = [tuple(e) for e in unlabeled[i]['relations']]
    gt = unlabeled[i]['relation_gts']
    gt = [tuple(e[:5]) for e in gt]
    print(pred)
    print(gt)
    if set(pred) != set(gt):
        no_match += 1
    print('========================')
```
